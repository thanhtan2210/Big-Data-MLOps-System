# 03. Two-Tower Model Prototyping

Notebook này thực hiện xây dựng và huấn luyện thử nghiệm mô hình Retrieval Two-Tower sử dụng TensorFlow/Keras.

```python
import tensorflow as tf
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Định nghĩa Kiến trúc Two-Tower

```python
EMBEDDING_DIM = 32

def create_user_tower(n_features):
    return tf.keras.Sequential([
        layers.Dense(128, activation='relu', input_shape=(n_features,)),
        layers.Dense(64, activation='relu'),
        layers.Dense(EMBEDDING_DIM, activation=None)
    ])

def create_movie_tower(n_features):
    return tf.keras.Sequential([
        layers.Dense(128, activation='relu', input_shape=(n_features,)),
        layers.Dense(64, activation='relu'),
        layers.Dense(EMBEDDING_DIM, activation=None)
    ])
```

## 2. Giả lập dữ liệu huấn luyện (Mocking Feature Store)

```python
num_samples = 5000
user_features_mock = np.random.rand(num_samples, 10)
movie_features_mock = np.random.rand(num_samples, 10)
labels = np.random.randint(2, size=num_samples)
```

## 3. Huấn luyện thử nghiệm

```python
user_tower = create_user_tower(10)
movie_tower = create_movie_tower(10)

# Loss function mô phỏng sự tương quan giữa 2 vector
user_tower.compile(optimizer='adam', loss='mse')
movie_tower.compile(optimizer='adam', loss='mse')

print("Bắt đầu huấn luyện mô hình User Tower...")
history = user_tower.fit(user_features_mock, np.random.rand(num_samples, EMBEDDING_DIM), epochs=5, verbose=1)

plt.plot(history.history['loss'])
plt.title("Training Loss")
plt.show()
```

## 4. Xuất Embeddings

```python
sample_user_emb = user_tower.predict(user_features_mock[:1])
print(f"Kích thước Vector Embedding của User: {sample_user_emb.shape}")
```
